```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
my_filepath = "/kaggle/input/olympic-country-efficiency-dataset-19602016/olympic_countries_efficiency.csv"
my_data = pd.read_csv(my_filepath)
#my_data[index_col=="# Year"]
```

```python
# Print the first five rows of the data
my_data.head()
```

```python
#Average population per year
plt.figure(figsize=(10,6))
plt.title("Average population per year")
plt.ylabel("Population")
sns.barplot(x=my_data["Year"], y=my_data["population"]) 
```

```python
#Group by country and income
my_data = my_data.rename(columns={"ISO3": "country"})
income_by_country = my_data.groupby(["country","income_group"])
income_by_country.head()
```

```python
plt.figure(figsize=(10,6))
plt.title("Number of events participated by income group")

sns.barplot(data=my_data, x="income_group", y="events_participated") 
```
